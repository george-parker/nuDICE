```python
# this code runs internally with eV so these might be useful
meter = 5.06773093741e6        # [eV^-1/m]
km = 1.0e3*meter               # [eV^-1/km]
MeV = 1.0e6                    # [eV/MeV]

# mass=="NO", nufit 6.1 w/ SK-atms

Delta_m21_sq = 7.537e-5        # [eV^2]
Delta_m31_sq = 2.511e-3        # [eV^2]
theta12 = 33.76*np.pi/180
theta13 = 8.62*np.pi/180
theta23 = 43.29*np.pi/180

U = Uall(theta12, theta23, theta13, 0)  # lepton mixing matrix (we set delta_CP=0)
```

```python
# === Neutrino mass spectrum (Normal Ordering) ===

# for mi -> mj in NO
def mass(m1):
    m2 = np.sqrt(Delta_m21_sq + sq(m1))
    m3 = np.sqrt(Delta_m31_sq + sq(m1))
    return m2, m3
```

```python
# PLOT PARAMS
g_s   = 0.05                                                # scalar coupling 
m1_   = 1e-3                                                # m1 mass in eV
dist  = 100*km                                              # propagation distance                
E_min = 0.                                                  # lowest energy bin lower edge
E_max = 5.*MeV                                              # highest energy bin upper edge
num   = 100                                                 # number of energy bins
```

```python
# calcuate the rest
g_ps  = g_s                                                 # we set g_ps = g_s
m2_, m3_ = mass(m1_)                                        # calculate masses of nu2, nu3
e_edges = np.linspace(E_min, E_max, num+1)                  # energy bin edges
e_centr = 0.5 * (e_edges[1:] + e_edges[:-1])                # energy bin centres
```

```python
# run the code
mass_array = []
flav_array = []
init_state = []

masses_list = np.array([m1_, m2_, m3_, 1e-1, 0.2, 0.5])

for N in [3, 4, 5, 6]:

    t0 = time.time()
    with Roll("System with "+str(N)+" neutrinos", N-1):
        
        masses = masses_list[:N]
    
        # turn on all decay channels i->j for i>j
        channel = {}
        for i in range(N):
            for j in range(i):
                channel[f"n{i+1}->n{j+1}"] = {"index_p": i, "index_d": j}

        # mixing matrix - extra flavors decouple
        U_N = np.zeros((N, N), dtype=np.complex128)
        U_N[:3, :3] = Uall(theta12, theta23, theta13, 0)
        for i in range(3, N):
            U_N[i, i] = 1.0                            

        # initial state: pure heaviest mass eigenstate for each configuration
        rho_mass = np.zeros((N, N), dtype=np.complex128)
        rho_mass[N-1, N-1] = 1.0    
        initial_value = rho_mass * np.ones(num).reshape(num, 1, 1)

        init_state.append(np.tile(np.real(np.diag(rho_mass)), (num, 1)))
        
        answer = dynam(initial_value, dist, e_edges, masses, g_s, g_ps, channel)
        mass_array.append(dibloc(answer))
        flav_array.append(dibloc([mass_to_flav(k, U_N, 0) for k in answer]))
```

```python
fig, axes = dice_plot(e_edges, mass_array, flav_array, init_state, g_s, dist, masses)

# plt.savefig("dice.pdf", dpi=1000)
# plt.show()
```

```python
# PLOT PARAMS
g_s   = 1e-2                                                # scalar coupling 
m1_   = 1e-3                                                # m1 mass in eV
dist  = 50*km                                               # propagation distance                
E_min = 0.                                                  # lowest energy bin lower edge
E_max = 3.*MeV                                              # highest energy bin upper edge
num   = 200                                                 # number of energy bins

# flat spectrum of pure electron ANTIneutrinos in the mass basis
mass_ebar = flav_to_mass([[1,0,0],
                          [0,0,0],
                          [0,0,0]],U,1) * np.ones(num).reshape(num, 1, 1)

# allowed decay channels (dual decay)
channel = {"n3->n2": { "index_p" : 2,"index_d" : 1},
           "n3->n1": { "index_p" : 2,"index_d" : 0}}
```

```python
# calcuate the rest
g_ps  = g_s                                                 # we set g_ps = g_s
m2_, m3_ = mass(m1_)                                        # calculate masses of nu2, nu3
masses = [m1_, m2_, m3_]
Ndim = len(masses)
e_edges = np.linspace(E_min, E_max, num+1)                  # energy bin edges
e_centr = 0.5 * (e_edges[1:] + e_edges[:-1])                # energy bin centres
```

```python
# load in an example reactor at 50 km
flux = np.loadtxt('reactor_flux.txt')[:,1]
```

```python
# now we extend our system to account for Majorana decays 
initial_maj = np.zeros((num, 2*Ndim, 2*Ndim), dtype=np.complex128)

# initial state: reactor spectrum of pure electron antineutrinos
initial_maj[:, Ndim:, Ndim:] = flux[:, None, None] * mass_ebar
masses_maj, channel_maj = make_majorana(masses, channel)  
```

```python
with Roll("Majorana decays at a reactor", 4):
    answer_maj = dynam(initial_maj, dist, e_edges, masses_maj, g_s, g_ps, channel_maj)
```

```python
# extract and combine the nu and antinu results
result_n = answer_maj[:, :Ndim, :Ndim]
result_a = answer_maj[:, Ndim:, Ndim:]

# mass basis
maj_n_mass = dibloc(result_n)
maj_a_mass = dibloc(result_a)
maj_mass_combined = np.concatenate([maj_n_mass, maj_a_mass], axis=-1)

# flav basis
maj_n_flav = dibloc([mass_to_flav(k, U, 0) for k in result_n])
maj_a_flav = dibloc([mass_to_flav(k, U, 1) for k in result_a])
maj_flav_combined = np.concatenate([maj_n_flav, maj_a_flav], axis=-1)
```

```python
maj_plot(e_edges, maj_flav_combined, maj_mass_combined, flux, g_s, dist, m1_)

# plt.savefig("majorana.pdf", dpi=1000)
# plt.show()
```
